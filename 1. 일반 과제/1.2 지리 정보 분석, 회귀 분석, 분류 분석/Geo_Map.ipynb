```python
import json
geo_path = '/content/skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
import pandas as pd
df = pd.read_csv('/content/seoul_wifi_zone.csv')
df.head()
```

```python
freq_gu = df['구명'].value_counts()
freq_gu
```

```python
map = folium.Map(location=[37.5102, 126.982], zoom_start=11, tiles='Stamen Toner')
map
```

```python
map.choropleth(geo_data = geo_str, 
               data = freq_gu, 
               colums = ['gu', 'counts'], 
               fill_color = 'PuRd',
               key_on = 'feature.id')
map
```
